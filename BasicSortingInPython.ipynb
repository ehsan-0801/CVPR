```python
def bubblesort(l):
    n=len(l)
    s=1
    for x in range(n-1):
        for y in range(0,n-x-1):
            if l[y]>l[y+1]:
                l[y],l[y+1] = l[y+1],l[y]
                print(f'Step {s}:',l,end="\n")
                s+=1
            else:
                print(f'Step {s}:',l,end="\n")
                s+=1
list1 = [int(item) for item in input("Enter the list items : ").split()]
#print(list1)
bubblesort(list1)
    
```

```python
def selectionsort(l2):
    n2=len(l2)
    s=1
    for i in range (n2):
        minidx = i
        for j in range (i+1,n2):
            if l2[j] < l2[minidx]:
                minidx = j
        l2[i], l2[minidx] = l2[minidx],l2[i]
        print(f'Step {s}:',l2,end="\n")
        s+=1
list1 = [int(item) for item in input("Enter the list items : ").split()]
#print(list1)
selectionsort(list1)
```

```python
def Insertionsort(l3):
    n3=len(l3)
    s=1
    for i in range (1,n3):
        kvalue = l3[i]
        j = i-1
        while j>=0 and kvalue<l3[j]:
            l3[j+1] = l3[j]
            j -= 1
            l3[j+1]= kvalue
        print(f'Step {s}:',l3,end="\n")
        s+=1
list1 = [int(item) for item in input("Enter the list items : ").split()]
#print(list1)
Insertionsort(list1)
```
